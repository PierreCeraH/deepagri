# Aggregating

## Custom aggregator

```python
def count_pos_outliers(series, threshold=1, greater=True):
    s_mean = series.mean()
    s_std = series.std()

    if greater:
        return (series >= (s_mean + threshold * s_std)).sum()
    elif not greater:
        return (series <= (s_mean - threshold * s_std)).sum()
    else:
        raise TypeError("Argument 'greater' should be a boolean")
```

```python
def count_neg_outliers(series, threshold=1, greater=False):
    s_mean = series.mean()
    s_std = series.std()

    if greater:
        return (series >= (s_mean + threshold * s_std)).sum()
    elif not greater:
        return (series <= (s_mean - threshold * s_std)).sum()
    else:
        raise TypeError("Argument 'greater' should be a boolean")
```
