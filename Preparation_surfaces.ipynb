```python
filepath = '/Users/pierre/code/PierreCeraH/deepagri/raw_data/cleaned/Histo_data_YieldsAreasProd.xlsx'

df = pd.read_excel(filepath)
df = df.rename(columns={'Unnamed: 0':'ANNEE'})

df
```

```python
df = df.loc[1:,:]
df = df.set_index('ANNEE')

cols = df.columns
cols_to_drop = []
for col in cols :
    if '1' in col:
        cols_to_drop.append(col)
    elif '2' in col:
        cols_to_drop.append(col)

df = df.drop(cols_to_drop, axis=1)
df = df.rename(columns={'TOTAL':'FRANCE'})

df_surfaces_France = df[['FRANCE']]

df = df.drop('FRANCE', axis=1)
df = df.loc[2010:,:]

cols = df.columns
dict_col_to_rename = {}
i=1
for col in cols :
    dict_col_to_rename[col]=i
    i = i+1
df = df.rename(columns=dict_col_to_rename)

dept_to_keep = [80,51,59,76,10,86,37,35,21,22,14,18,61,17,56,54,29,44,50,63,67,31,68,26,12,43,87,69,46,9,33,65,40,5,34,13,83,66]
cols_to_drop = []

for i in range(1,96):
    if i not in dept_to_keep :
        cols_to_drop.append(i)
df = df.drop(columns=cols_to_drop)

df_surfaces = pd.DataFrame(df.stack())

df_surfaces = df_surfaces.rename(columns={0:'Production'})
df_surfaces = df_surfaces.reset_index()
df_surfaces['Ind']=df_surfaces['ANNEE'].astype(str)+'-'+df_surfaces['level_1'].astype(str)

df_surfaces = df_surfaces.drop(['level_1','ANNEE'], axis=1)
df_surfaces = df_surfaces.set_index('Ind')
```
